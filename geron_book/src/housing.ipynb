```python
def test_set_check(id, ratio, hash):
    return hash(np.int64(id)).digest()[-1] < 256 * ratio

def split_data_by_id(data, ratio, id_column, hash=hashlib.md5):
    ids = data[id_column]
    in_test_set = ids.apply(lambda id_ : test_set_check(id_, ratio, hash))
    return data.loc[~in_test_set], data.loc[in_test_set]
```

```python
df_ids = df.reset_index()
```

```python
df_ids.head()
```

```python
train_set, test_set = split_data_by_id(df_ids, 0.2, "index")
```

```python
df["income_cat"] = np.ceil(df["median_income"] / 1.5)
df["income_cat"].where(df["income_cat"] < 5, 5.0, inplace=True)
```

```python
df["income_cat"].hist()
```

```python
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(df, df["income_cat"]):
    strat_train_set = df.loc[train_index]
    strat_test_set = df.loc[test_index]
```

```python
df["income_cat"].value_counts() / len(df)
```

```python
for set_ in (strat_train_set, strat_test_set):
    set_.drop("income_cat", axis=1, inplace=True)
```

```python
df_train_copy = strat_train_set.copy()
```
